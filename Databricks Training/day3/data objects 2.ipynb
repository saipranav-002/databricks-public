## Types of views

A virtual table 
1. standard view ( persisted save to schema)

```python
%sql
create view acc_dev.pranav.top_customers as 
select customer_id,sum(total_amount) as total_amount from acc_dev.pranav.sales group by customer_id sort by total_amount desc limit 3
```

```python
%sql
create temp view last_custom_view as 
select  customer_id,sum(total_amount) as total_amount from acc_dev.pranav.sales group by customer_id sort by total_amount limit 10
```

```python
%sql
create function acc_dev.pranav.full_name(first_name string, last_name string)
returns string
return concat(first_name,' ',last_name);
    
select acc_dev.pranav.full_name('sai','kee')
```
