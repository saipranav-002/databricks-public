### End to End ETL for financials

```python
%skip
Task.

step 1: create a dataframe from this path: /Volumes/acc_dev/naval/raw/financial_dataset.csv
Step 2: rename the all cols and make it lower case
        ii. add new col for auditing: (ingestion date)
        iii. add new col for path: (input_path)
Step 3: save final df to table       
note: keep it clean Modular code
```

```python
from pyspark.sql.functions import *
```

```python
input_path="/Volumes/acc_dev/naval/raw/financial_dataset.csv"
```

```python
df=spark.read.csv(input_path,header=True,inferSchema=True)
```

```python
%skip
df.columns
```

```python
new_col=['transaction_id',
 'customer_id',
 'account_type',
 'transaction_type',
 'amount_usd',
 'branch',
 'transaction_date']
```

```python
df1=df.toDF(*new_col).withColumn("ingestion_date", current_timestamp()).withColumn("path",col("_metadata.file_name"))
```

```python
df1.write.mode("overwrite").saveAsTable("acc_dev.naval.financials")
```
